# Benchmark of ODE Solvers Using a 1D ODE With Changing Mass

This page shows the results of a benchmark of ODE solvers using a 1D ODE with changing mass.


## Problem

### Equation

$$
e^y \dot{y} = e^y y
$$

### Initial Condition

$$
y(0) = 1
$$

### Notes

- This problem results in $\dot{y} = y$, which has the exact solution $y(t) = e^t$.


## Results


```python
import num_collect_docs_utils.plot_common

num_collect_docs_utils.plot_common.set_common_configuration()
```

```python
# Load data
import gzip
import msgpack
import pandas

with gzip.open("data/changing_mass_exponential_problem.data", mode="rb") as file:
    results = msgpack.unpack(file)
data_frame = pandas.DataFrame(
    {
        "Solver": results["solver_list"],
        "Err. Tol.": results["tolerance_list"],
        "Error Rate": results["error_rate_list"],
        "Time [sec]": results["time_list"],
    }
)
```

```python
# Show plots
import plotly.express
import num_collect_docs_utils.plot_common

figure = plotly.express.line(
    data_frame,
    x="Time [sec]",
    y="Error Rate",
    color="Solver",
    line_dash="Solver",
    hover_data=["Solver", "Err. Tol.", "Error Rate", "Time [sec]"],
    markers=True,
    log_x=True,
    log_y=True,
    title="Work-Error Diagram of ODE Solvers for 1D ODE With Changing Mass",
)
num_collect_docs_utils.plot_common.show_figure(figure)
```
